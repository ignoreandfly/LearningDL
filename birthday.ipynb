# Birthday problem

### If there are 'n' people in a room, for what value of 'n' is the probability that at least two people sharing the same birthday is > 50%

```python
n = 1
curr_prob = 1
curr_prob_no_share = []

while True:
     curr_prob *= (365-n)/365
     curr_prob_no_share.append(curr_prob)
     n +=1
     if (1-curr_prob > 0.5):
          break

curr_prob_no_share = [ 1- prob for prob in curr_prob_no_share]
```

#### Plot the probability of at least two people sharing birthday vs n

```python
import matplotlib.pyplot as plt

plt.scatter(list(range(n+1)[2:]), curr_prob_no_share, color='b', label='P(at least two share birthday)')
plt.xlabel('Number of People in the Room (n)')
plt.ylabel('Probability')
plt.title('Probability of Shared Birthdays vs. Number of People')
plt.grid(True)
plt.xticks(range(min(list(range(n+1)[2:])), max(list(range(n+1)[2:])) + 1, 1))  
plt.legend()
plt.show()
```
